```python
import pandas as pd
import random
```

```python
df = pd.read_csv("Truck_data_Labeled_random_leaders.csv")
df.head()
```

```python
MaxNumberOfData = 5000

def GenerateData(random_min, random_max):
    return [random.randint(random_min, random_max) for _ in range(MaxNumberOfData)]
```

```python
distance = GenerateData(500, 1000)
route_match = GenerateData( 0, 100)
fuel_consumption = GenerateData( 30, 150)
engine_temperature = GenerateData( 36, 80)
body_weight = GenerateData( 1200, 4000)
sensor_equipped = GenerateData( 0, 1)
```

```python
df['distance'] = distance
df['route_match'] = route_match
df['fuel_consumption'] = fuel_consumption
df['engine_temperature'] = engine_temperature
df['body_weight'] = body_weight
df['sensor_equipped'] = sensor_equipped
```

```python
df.to_csv('DataCreation.csv')
```
